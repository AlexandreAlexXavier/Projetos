```python
#Criando a coluna de receita
df['Receita'] = df['Vendas'].mul(df['Qtde'])
```

```python
#Transformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#Transformando coluna de data em data
df['Data'] = pd.to_datetime(df['Data'])
```

```python
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df['Data'].dt.year)['Receita'].sum()
```

```python
#Criando e exibindo uma nova coluna com o ano
df['Ano_Venda'] = df['Data'].dt.year
df.sample(5)
```

```python
#Extraindo o mês e o dia
df['mes_venda'], df['dia_venda'] = (df['Data'].dt.month, df['Data'].dt.day)
df.sample(5)
```

```python
#Retornando a data mais antiga
df['Data'].min()
```

```python
#Calculando a diferença de dias
df['diferenca_dias'] = df['Data'] - df['Data'].min()
df.sample(5)
```

```python
#Criando a coluna de trimestre
df['trimestre_venda'] = df['Data'].dt.quarter
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 3)]
vendas_marco_19.sample(10)
```

```python
#Criando nosso Dataframe
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a Receita Total?
df['Valor Venda'].sum()
```

```python
#Qual o custo total?
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df.head(1)
```

```python
#Qual o custo Total?
df['Custo'].sum()
```

```python
#Cálculo Lucro Total
df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
#Total Lucro
round(df['Lucro'].sum(),2)
```

```python
#Total de dias para enviar o produto
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
#Extraindo apenas os dias
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.dtypes
```

```python
#Médio do tempo de envio por Marca
round(df.groupby('Marca')['Tempo_envio'].mean(),2)
```

```python
#Verificando dados faltantes
df.isnull().sum()
```

```python
#Agrupamento por Ano e Marca
pd.options.display.float_format = '{:20,.2f}'.format
df.groupby([df['Data Venda'].dt.year,'Marca'])['Lucro'].sum()
```

```python
#Total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total Produtos Vendidos', color='green')
plt.xlabel('Total de vendas')
plt.ylabel('Produto');
```

```python
#Lucro por ano
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('Lucro');
```

```python
#Vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
df_2009.head()
```

```python
#Gráfico Lucro x Mês em 2009
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucro x Mês', marker = 'v')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
#Gráfico Lucro x Marca em 2009
df_2009.groupby(df_2009['Marca'])['Lucro'].sum().plot.bar(title='Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
